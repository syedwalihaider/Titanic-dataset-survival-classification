```python
df.dtypes[df.dtypes==object]
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler,OrdinalEncoder,OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1,stratify=y)
```

```python
transformer = ColumnTransformer(transformers=[('scaling',StandardScaler(),["age","fare"]),
                                              ('ordinal encoding',OrdinalEncoder(categories=[["female","male"]]),["sex"]),
                                              ('nominal encoding',OneHotEncoder(sparse=False,drop='first'),["cabin","embarked"])],
                                remainder='passthrough')
```

```python
x_tr_train = transformer.fit_transform(x_train)
```

```python
x_tr_train.shape
```

```python
y_train.shape
```

```python
x_tr_test = transformer.transform(x_test)
x_tr_test.shape
```

```python
clf = GridSearchCV(LogisticRegression(),{"C":[0.1,1,10],"class_weight":["balanced","None"]},cv=5,scoring='f1')
```

```python
clf.fit(x_tr_train,y_train)
```

```python
pd.DataFrame(clf.cv_results_)
```

```python
y_test_pred = clf.predict(x_tr_test)
```

```python
y_train_pred = clf.predict(x_tr_train)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_train_pred,y_train))
```

```python
print(classification_report(y_test_pred,y_test))
```

```python
clf1 = GridSearchCV(KNeighborsClassifier(),{"n_neighbors":[9,11,13],"p":[1,2]},cv=5,scoring='f1')
```

```python
clf1.fit(x_tr_train,y_train)
y_train_pred = clf1.predict(x_tr_train)
y_test_pred = clf1.predict(x_tr_test)
```

```python
pd.DataFrame(clf1.cv_results_)
```

```python
print(classification_report(y_train_pred,y_train))
print(classification_report(y_test_pred,y_test))
```

```python
clf2 = GridSearchCV(GaussianNB(),{"var_smoothing":[2e-01,1e-01,1e-03]},cv=5,scoring='f1')
clf2.fit(x_tr_train,y_train)
y_train_pred = clf2.predict(x_tr_train)
y_test_pred = clf2.predict(x_tr_test)
pd.DataFrame(clf2.cv_results_)
```

```python
print(classification_report(y_train_pred,y_train))
print(classification_report(y_test_pred,y_test))
```

```python
clf3 = GridSearchCV(SVC(),{"C":[0.1,1,10],"kernel":["rbf",'linear'],"gamma":["auto","scale"],"class_weight":["balanced"]},cv=5,scoring='f1')
clf3.fit(x_tr_train,y_train)
y_train_pred = clf3.predict(x_tr_train)
y_test_pred = clf3.predict(x_tr_test)
pd.DataFrame(clf3.cv_results_)
```

```python
print(classification_report(y_train_pred,y_train))
print(classification_report(y_test_pred,y_test))
```

```python
clf4 = GridSearchCV(DecisionTreeClassifier(),{"max_depth":[1,5,8,13],"class_weight":["balanced"]},cv=5,scoring='f1')
clf4.fit(x_tr_train,y_train)
y_train_pred = clf4.predict(x_tr_train)
y_test_pred = clf4.predict(x_tr_test)
pd.DataFrame(clf4.cv_results_)
```

```python
print(classification_report(y_train_pred,y_train))
print(classification_report(y_test_pred,y_test))
```

```python
clf5 = GridSearchCV(RandomForestClassifier(),{"max_depth":[4,5,6,7],"class_weight":["balanced"],"n_estimators":[150,200]},cv=5,scoring='f1')
clf5.fit(x_tr_train,y_train)
y_train_pred = clf5.predict(x_tr_train)
y_test_pred = clf5.predict(x_tr_test)
pd.DataFrame(clf5.cv_results_)
```

```python
print(classification_report(y_train_pred,y_train))
print(classification_report(y_test_pred,y_test))
```
